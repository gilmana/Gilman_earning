# Part #1

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(xlim = [0,10], ylim = [-2,8], title = "Example Title",
       ylabel = "Y-lab", xlabel = "X-lab")
```

```python
x = np.linspace(0,(2*3.14)*2,1000)
```

```python
y = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows = 3 , ncols = 1, figsize = (5,4))

ax[0].plot(x,y)
ax[0].set(title = "Signal 1", xticks = [], yticks = [])
ax[1].plot(x, np.cos(x + (3.14/4)))
ax[1].set(title = "Signal 2", xticks = [], yticks = [])
ax[2].plot(x, np.cos(x + 3.13/2))
ax[2].set(title = "Signal 3", xticks = [], yticks = [])

plt.tight_layout()
```

```python
np.random.seed(1)

y = np.random.randn(100).cumsum()
x = np.linspace(0,10,100)

fig, axe = plt.subplots()
axe.plot(x,y)
axe.fill_between(x,y, color = "lightblue")
```

```python
x = np.linspace(0,10,200)
y1 = 2*x +1
y2 = 3*x +1.2
y_mean = 0.5*x + np.cos(x*2) + 2.5*x + 1.1

fig, axe = plt.subplots()

axe.plot(x,y1)
axe.plot(x,y2)
axe.plot(x,y_mean)

axe.fill_between(x,y1,y2, color = "yellow")
```
