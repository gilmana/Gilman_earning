```python
import numpy as np
import matplotlib.pyplot as plt



%matplotlib inline
```

# Part #1

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(xlim = [0,10], ylim = [-2,8], title = "Example Title",
       ylabel = "Y-lab", xlabel = "X-lab")
```

```python
x = np.linspace(0,(2*3.14)*2,1000)
```

```python
y = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows = 3 , ncols = 1, figsize = (5,4))

ax[0].plot(x,y)
ax[0].set(title = "Signal 1", xticks = [], yticks = [])
ax[1].plot(x, np.cos(x + (3.14/4)))
ax[1].set(title = "Signal 2", xticks = [], yticks = [])
ax[2].plot(x, np.cos(x + 3.13/2))
ax[2].set(title = "Signal 3", xticks = [], yticks = [])

plt.tight_layout()
```

```python
np.random.seed(1)
x = np.arange(5)
y = np.random.randn(5)
```

```python
fig, axes = plt.subplots(ncols = 2, figsize = plt.figaspect(1/2))
axes[0].bar(x,y,)
axes[1].barh(x,y)

#draw horizontal line

axes[0].axhline(0, color= "black", linewidth = 1)
axes[1].axvline(0, color = "black", linewidth = 1)
```
