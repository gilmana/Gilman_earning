### Using Dave's code for pearson correlation as an example

```python
#Dave's code as a stencil to understand and follow: 
def compare_all_rows_pearson(df):
    """df is a pandas dataframe
        returns a symmetrical metric matrix"""
    
    metric = pd.DataFrame()
    
    for index_i, row_i in df.iterrows():
        for index_j, row_j in df.iterrows():
            if index_j < index_i:
                continue
            if index_j == index_i:
                metric.loc[index_i, index_j] = 1
                continue
            metric.loc[index_i, index_j] = row_i.corr(row_j, 'pearson')
            metric.loc[index_j, index_i] = metric.loc[index_i, index_j]
            
    return metric
```

```python
# the code below, creates a generator that has both index and series. I print the index and then the sum in the Series
# good stuff! 
for i,j in df_values_only.iterrows():
    print(i, sum(j))
```

```python
#Ok lets increase complexity - lets compare the difference between the sum of the first row and all remaining rows. 
n = 0 
for i, j in df_values_only.iterrows():
    while n < 1: #limiting to the first for loop to the first row
        n += 1
        for x, y in df_values_only.iterrows(): # comparing all other rows to the first row
            print(i, x, sum(j) - sum(y))
```

```python
#if we remove the restriction of the first row, we will get all the combinations of rows for the nested for loops. 
for i, j in df_values_only.iterrows():
    for x, y in df_values_only.iterrows(): # comparing all other rows to the first row
        print(i, x, sum(j) - sum(y))
```

```python
#now we need to combine this nested for loop to create a dataframe and insert the right values in the indices. 

metric = pd.DataFrame()

for index_i, row_i in df_values_only.iterrows():
    for index_j, row_j in df_values_only.iterrows():
        metric.loc[index_i, index_j] = abs(sum(row_i)-sum(row_j))

#this works, and its beautiful, and I have a complete understanding of it. lets just make a function out of it. 
```

```python
def Difference_in_row_sums(df):
    
    metric = pd.DataFrame()

    for index_i, row_i in df.iterrows():
        for index_j, row_j in df.iterrows():
            metric.loc[index_i, index_j] = abs(sum(row_i)-sum(row_j))


    return metric
```

### Understanding python generators 

https://wiki.python.org/moin/Generators

http://stackoverflow.com/questions/231767/what-does-the-yield-keyword-do-in-python


The understanding hierarchy: iterators < generators < yield

iteratorables just performing a repetition over items in a list - iteratrate over every item. 

generators are iterators which you can only iterate over once. They generate the values on the fly, and do not 
store them in memory. 

Lets look at both... 


```python
#lets make a simple list and iterate an operation over the list
my_list = [0,1,2,3]
for i in my_list:
    print (i**2)
```

```python
#can create a list using list comprehension and iterate over that list the same way 
my_list = [x*x for x in range(4)]
for i in my_list:
    print(i)
```

```python
#We can create a generator the same way that we created a list comprehension but using () instead of []
#remember, this generates the values only when it needs to (called upon) on the fly, then forgets it
my_generator = (x*x for x in range(4))
for i in my_generator:
    print(i)
```

```python
# The difference is noticed if you try to run the for loop a second time... it wont work, 
# generators calculate each value once when called upon and then forget
for i in my_generator:
    print(i)
```

OK, so whats the deal with Yield? 

Yield is a keyword similar to return, except the function will return a genertor. Interesting... 

```python
def new_generator():
    my_range = range(4)
    for i in my_range:
        yield i*i

my_generator = new_generator()
```

```python
#this is a generator object!!!! 
type(my_generator)
```

```python
#lets iterate over the generator object! 
for i in my_generator:
    print(i)
```

words of wisdom:

"To master yield, you must understand that when you call the function, the code you have written in the function body does not run. The function only returns the generator object, this is a bit tricky :-)"
