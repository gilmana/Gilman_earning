```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Exploring pandas plot wrapper

```python
x = np.arange(20)

A = x*(1.5)+2
B = x*(-1.5)-2
C = (x-5)
D = (x)*0.5

arr_list = [A,B,C,D]
```

```python
stacked = np.stack(arr_list, axis = 1)
df = pd.DataFrame(stacked)
df.columns = ["A","B","C","D"]
```

```python
df.plot(kind = "line", marker = "o", ls = "None")
```

```python
df.plot(kind = "bar")
```

```python
df.plot(kind = "hist", alpha = 0.5, subplots = True)
```

```python
df.plot(kind = "box")
```

```python
df.plot(kind = "kde")
```
