# merge_asof
Merge based on how close the values are. Useful for merging based on datetime values for multiple data sources. A left Join is performed. 

```python
merge_asof = pd.merge_asof(inst1,inst2,
                          on = "date/time",
                          direction = "nearest",
                          tolerance = pd.Timedelta("30 minutes"))

merge_asof
```

# merge_asof followed by an outer merge on a scaffold column
Allows for retaining of all the data that is outside the tolerance time.

```python
inst2["Rank"] = inst2["date/time"].rank()

merge_asof = pd.merge_asof(inst1,inst2,
                          on = "date/time",
                          direction = "nearest",
                          tolerance = pd.Timedelta("30 minutes"))

merge_asof.merge(inst2, on = "Rank", how = "outer")
```

As you can see, there are duplicate columns from inst2 (because of a second merge which was outer on rank). It could be simpler to delete all inst2 columns from the first merge (merge_asof) except for Rank column. This way we can avoid confusion from duplicate columns - this is just a scaffold step to do the real merge on the Rank column. 

```python
inst2["Rank"] = inst2["date/time"].rank()

merge_asof = pd.merge_asof(inst1,inst2,
                          on = "date/time",
                          direction = "nearest",
                          tolerance = pd.Timedelta("30 minutes"))

#merge_asof.merge(inst2, on = "Rank", how = "outer") #will do outer join after dropping columns from inst2
```

Another way of doing the same drop of columns. This is a simpler way, Ill go with it.

Now just need to clean up:
    - create one master date/time column
    - create on master sample ID column
    - drop scaffolds (Rank, instrument specific Sample ID, instrument specific date/time)

```python
merge_asof.filter(like = "date/time") #filter all columns for similar to "date/time" str
merge_asof.filter(like = "date/time").bfill(1) # also backfill values along axis 1 
merge_asof.filter(like = "date/time").bfill(1).iloc[:,0] # now only select the first column

merge_asof = merge_asof.assign(timestamp = lambda x: x.filter(like = "date/time").bfill(1).iloc[:,0]) # assign the result a new column "timestamp"

merge_asof = merge_asof.assign(Name = lambda x: x.filter(like = "Sample ID").bfill(1).iloc[:,0]) #same thing for sample Name 
```

```python
# Drop all unnecessary columns 
merge_asof.drop(merge_asof.filter(like = "date/time").columns,axis = 1, inplace =True)
merge_asof.drop(merge_asof.filter(like = "Sample ID").columns,axis = 1, inplace =True)
merge_asof.drop(["Rank"], axis = 1, inplace = True)
```

#### Combining it all together

```python
inst2["Rank"] = inst2["date/time"].rank() #scaffold column

#merge on timestamp
merge_asof = pd.merge_asof(inst1,inst2,
                          on = "date/time",
                          direction = "nearest",
                          tolerance = pd.Timedelta("30 minutes"))

#deleting all but Rank from inst2 in prep for outer merge
drop_ = inst2.columns[~inst2.columns.isin(["Rank", "date/time"])]
merge_asof.drop(drop_, axis = 1, inplace = True)

#Now can perform the outer merge
merge_asof = merge_asof.merge(inst2, on = "Rank", how = "outer")

#Backfill the main identifier columns from both instruments into into single columns
merge_asof = merge_asof.assign(timestamp = lambda x: x.filter(like = "date/time").bfill(1).iloc[:,0]) # assign the result a new column "timestamp"
merge_asof = merge_asof.assign(Name = lambda x: x.filter(like = "Sample ID").bfill(1).iloc[:,0]) #same thing for sample Name 

# Drop all unnecessary columns 
merge_asof.drop(merge_asof.filter(like = "date/time").columns,axis = 1, inplace =True)
merge_asof.drop(merge_asof.filter(like = "Sample ID").columns,axis = 1, inplace =True)
merge_asof.drop(["Rank"], axis = 1, inplace = True)

# Final rearanging of columns
cols = ['Name','timestamp', 'Viable cells/ml (x10^6)', 'Viability(%)' ,'Gln', 'Glu', 'Gluc', 'Lac',
        'NH4+', 'Na+', 'K+', 'Ca++']
df = merge_asof[cols]

#Done
df
```
